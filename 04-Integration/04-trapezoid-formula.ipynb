### Формула трапеций

_Рассмотрим интеграл:_ $\quad I = \int\limits _a^b f(x)dx$

$$I \approx \sum\limits _{i=0}^{N-1} (x_{i+1} - x_i) \: \frac{f(x_i) + f(x_{i+1})}{2}$$

_Если узлы на равном расстоянии $h$ друг от друга:_

$$I \approx \sum\limits _{i=0}^{N-1} \frac{h}{2} \big(f(x_i) + f(x_{i+1})\big) = \frac{h}{2}\Big( \: f(a) + 2\sum\limits _{i=1}^{N-1}f(x_i) + f(b) \: \Big), \quad h = \frac{b-a}{N}$$


Рассматриваем функцию:

$$f(x) = \text{sin }x + \frac{1}{2}\text{cos }x, \qquad x \in [a, b]$$

```python
import numpy as np
from scipy.integrate import quad
```

```python
def f(x):
    return np.sin(x) + 0.5*np.cos(x)

a, b = -1, 2
h = 0.1

nodes = np.arange(a, b + h, h)
N = len(nodes)

y = f(nodes)

print('Number of nodes =', N)
```

```python
I = h/2 * (y[0] + 2*np.sum(y[1:][:-1]) + y[-1]) # аппроксимация с шагом h
print(I)
```

- Для вычисления определенного интеграла еще есть функция [`quad`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html)
    - возвращает значения интеграла и оценку абсолютной погрешности в результате

```python
I_q, abs_error = quad(f, a, b)

print("Integral =", I_q)
print("Absolute error =", abs_error)
```

_Первообразная функции_ $f(x)$:

$$F(x) = -\text{cos }x + \frac{1}{2}\text{sin }x$$

```python
def F(x):
    return -np.cos(x) + 0.5*np.sin(x)

I_exact = F(b) - F(a)
err = np.abs(I - I_exact)

print(I_exact)
print("err =", err)
```

### Оценка ошибки

$$\varepsilon = \frac{(b-a)^3}{12N^2} f''(\eta(x^*)) = \frac{h^2}{12}(b-a) f''(\eta(x^*)), \qquad x^* \in [a, b]$$

$I_{(1)}, \: I_{(2)}$ - приближения интеграла по двум схемам с шагами $h_{(1)}, \: h_{(2)}$ соответственно

$$I_{(2)} - I_{(1)} \approx C \: \frac{b-a}{12}  \: \big( h^2_{(1)} - h^2_{(2)} \big), \quad C \approx f''(\eta(x^*))$$

$$C \approx \frac{12}{b-a} \: \frac{I_{(2)} - I_{(1)}}{h^2_{(1)} - h^2_{(2)} }$$

```python
# более грубая аппроксимация с шагом 2h
I_ = h * (y[0] + 2*np.sum(y[::2][1:][:-1]) + y[-1])

C = 12/(b-a) * (I_ - I)/(h**2 - (2*h)**2)
err_est = C * ((b-a)**3)/(12*N**2)

print(err, err_est)
```
